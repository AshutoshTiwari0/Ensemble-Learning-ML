Ensemble Learning is the science of using diff models(weak learners) and using each one of their understanding to get an overall result. 
For example-> Instead of asking only a single person for an opinion we tend to ask many people and then overall come to a conclusion.

1.Bagging (Bootstrap Aggregating): This method involves training multiple instances of the same algorithm on different subsets of the training data (often created through bootstrapping) and then averaging the predictions. Random Forest is a well-known example of bagging.

2.Boosting: In this method, models are trained sequentially, with each new model focusing on correcting the errors made by the previous ones. Popular boosting algorithms include AdaBoost, Gradient Boosting, and XGBoost.

3.Stacking: This technique involves training multiple models and then using another model, often called a "meta-learner," to combine their predictions. The meta-learner tries to learn the best way to combine the outputs of the base models.

4.Voting: This is a simpler method where multiple models are trained and their predictions are combined by voting. The final prediction is typically the one that receives the majority of votes (for classification) or the average prediction (for regression).

<h1>ADABOOST (adapt to prev models mistakes)</h1>

Generate several models 
see first models misclassification. In next iteration inc the misclassified weights so that algorithm focuses more on misclassified data points.

GPT DEFINITION
Initial Model: AdaBoost starts by training a weak learner (often a decision stump) on the dataset.
<br>
Misclassification: After the first iteration, it identifies the misclassified samples.
<br>
Adjust Weights: The algorithm increases the weights of the misclassified data points (<b>and reduce weight of correct ones so that overall sum is same</b>) so that in the next iteration, the weak learner focuses more on these harder-to-classify points.
<br>
New Model: A new weak learner is trained with the adjusted weights, giving more attention to the previously misclassified points.
<br>
Repeat: This process is repeated for a specified number of iterations, with each subsequent learner trying to correct the errors of the previous ones.
<br>
Final Model: The final model is a weighted combination of all the weak learners.
